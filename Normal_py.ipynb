```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file=np.loadtxt("normaldata1.txt")
print(file)
```

```python
x1=file[:, 0]
y1=file[:,2]
y2=file[:,5]
y3=file[:,8]
plt.xlabel('Number of Buyers')
plt.ylabel('Total payment')
plt.title('Buyers Payment in QUAD, McAfee and PPM')
width = 1.20
plt.bar(x1,y2,
        width = width, 
        label='QUAD',fill=False, hatch='///',facecolor='lightgray',edgecolor='darkgreen')
plt.bar(x1+width,y1,
        width = width, 
        label='McAfee',fill=False,hatch='....',facecolor='lightgray',edgecolor='red')

plt.bar(x1+width+width,y3,
        width = width,
        label='PPM',fill=False, hatch='\\\\\\' ,facecolor='lightgray' ,edgecolor='blue')

plt.legend()
  
plt.show()
```

```python
x1=file[:, 1]
y1=file[:,3]
y2=file[:,6]
y3=file[:,9]
plt.xlabel('Number of Sellers')
plt.ylabel('Total payment')
plt.title('Sellers Payment in QUAD, McAfee and PPM')
width = 1.20
plt.bar(x1,y2,
        width = width, 
        label='QUAD',fill=False, hatch='///',facecolor='lightgray',edgecolor='darkgreen')
plt.bar(x1+width,y1,
        width = width, 
        label='McAfee',fill=False,hatch='....',facecolor='lightgray',edgecolor='red')

plt.bar(x1+width+width,y3,
        width = width,
        label='PPM',fill=False, hatch='\\\\\\' ,facecolor='lightgray' ,edgecolor='blue')
plt.legend()
  
plt.show()
```

```python
x1=file[:,11]
y1=file[:,12]
y2=file[:,13]
y3=file[:,14]
plt.plot(x1,y1,label="McAfee")
plt.plot(x1,y2,label="QUAD")
plt.plot(x1,y3,label="PPM")
plt.xlabel('Number of traders')
plt.ylabel('GFT Value')
plt.legend()
plt.show()
```
